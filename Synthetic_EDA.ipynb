```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming your data is stored in a CSV file named 'your_dataset.csv'
df = pd.read_csv('Synthetic_data_EDA_Sheet.csv')

# Display basic statistics of the dataset
print(df.describe())

# Check for missing values
print(df.isnull().sum())

# Visualize the distribution of numerical features
numerical_features = ['Timestamp']
sns.histplot(data=df[numerical_features])
plt.title('Distribution of Timestamp')
plt.show()

# Visualize the correlation matrix for numerical features
correlation_matrix = df[numerical_features].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()

# Visualize the count of categorical features
categorical_features = ['DeviceName', 'Weather', 'Activity', 'RoomName', 'PowerStatus', 'Week', 'UserId', 'Routines']

for feature in categorical_features:
    plt.figure(figsize=(10, 6))
    sns.countplot(data=df, x=feature, order=df[feature].value_counts().index)
    plt.title(f'Count of {feature}')
    plt.xticks(rotation=45)
    plt.show()
```

## Convert Each device minutes into hours

Then 1 more column as hours/minutes of each device, then we can recommend what time we can use what device.. 

```python

# Convert 'Timestamp' column to datetime format
df['Timestamp'] = pd.to_datetime(df['Timestamp'])

# Extract hours and minutes from the 'Timestamp' column
df['Hours'] = df['Timestamp'].dt.hour
df['Minutes'] = df['Timestamp'].dt.minute

# Added Hours column to existing df 
model_df = df.copy()

# Display the updated DataFrame
print(model_df)
```

```python
model_df.tail(10)
```

```python

# Group by 'DeviceName' and 'Day' and calculate the average usage time
usage_time = model_df.groupby(['DeviceName', model_df['Timestamp'].dt.day_name()])['Timestamp'].mean()

# Display the recommended usage time for each device on each day
print(usage_time)
```

```python

# Extract the day of the week from the timestamp
model_df['DayOfWeek'] = model_df['Timestamp'].dt.day_name()

# Separate patterns for Monday to Friday, Saturday, and Sunday
weekday_pattern = model_df[model_df['DayOfWeek'].isin(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday'])]
saturday_pattern = model_df[model_df['DayOfWeek'] == 'Saturday']
sunday_pattern = model_df[model_df['DayOfWeek'] == 'Sunday']

# Display the patterns for each group
print("Weekday Pattern:")
print(weekday_pattern['DeviceName'].value_counts())

print("\nSaturday Pattern:")
print(saturday_pattern['DeviceName'].value_counts())

print("\nSunday Pattern:")
print(sunday_pattern['DeviceName'].value_counts())
```

```python
import pandas as pd
import numpy as np

# Convert 'Timestamp' to datetime format
model_df['Timestamp'] = pd.to_datetime(model_df['Timestamp'])

# Extract day of the week and hour of the day
model_df['day_of_week'] = model_df['Timestamp'].dt.day_name()
model_df['hour_of_day'] = model_df['Timestamp'].dt.hour

# Group by 'day_of_week', 'hour_of_day', and 'DeviceName' to get the count of devices used at each time slot
time_device_count = model_df.groupby(['day_of_week', 'hour_of_day', 'DeviceName']).size().reset_index(name='count')

# Pivot the table to have 'day_of_week' and 'hour_of_day' as index, 'DeviceName' as columns, and 'count' as values
time_device_pivot = time_device_count.pivot_table(index=['day_of_week', 'hour_of_day'], columns='DeviceName', values='count', fill_value=0)

# Calculate the correlation matrix to identify patterns and similarities
correlation_matrix = time_device_pivot.corr()

# Print the correlation matrix
print("Correlation Matrix:")
print(correlation_matrix)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

# Assuming your dataset is named 'df'
# Features: 'Time', 'RoomName', 'Week', 'Hours', 'Minutes'
# Target: 'DeviceName'

# Convert 'Time' to datetime
df['Time'] = pd.to_datetime(df['Time'])

# Extract hours from 'Time' and create a new column 'Hour'
df['Hour'] = df['Time'].dt.hour

# Select relevant features and target
features = ['Hour', 'RoomName_Bathroom', 'RoomName_Bedroom', 'RoomName_Kitchen', 'RoomName_LivingRoom',
            'Week_Friday', 'Week_Monday', 'Week_Saturday', 'Week_Sunday', 'Week_Thursday', 'Week_Tuesday', 'Week_Wednesday',
            'Hours', 'Minutes']
target = 'DeviceName'

# Drop rows with missing values
df = df.dropna(subset=features + [target])

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df[features], df[target], test_size=0.2, random_state=42)

# Initialize the decision tree classifier
classifier = DecisionTreeClassifier()

# Train the classifier
classifier.fit(X_train, y_train)

# Test the classifier
accuracy = classifier.score(X_test, y_test)
print(f"Accuracy: {accuracy:.2f}")

```
